```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.xlabel("x-axis",fontsize=20)
plt.ylabel("y-axis",fontsize=20)
plt.title("Step Plot",fontsize=30)

plt.step(x,y)
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.xlabel("x-axis",fontsize=20)
plt.ylabel("y-axis",fontsize=20)
plt.title("Step Plot",fontsize=30)

plt.step(x,y,marker="o")
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.xlabel("x-axis",fontsize=20)
plt.ylabel("y-axis",fontsize=20)
plt.title("Step Plot",fontsize=30)

plt.step(x,y,marker="o",ms=10,mfc="r")    # ms= marker size, mfc= marker font,color
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.xlabel("x-axis",fontsize=20)
plt.ylabel("y-axis",fontsize=20)
plt.title("Step Plot",fontsize=30)

plt.step(x,y,marker="o",ms=10,mfc="r",label="Python")
plt.grid()
plt.legend()
plt.show()
```
