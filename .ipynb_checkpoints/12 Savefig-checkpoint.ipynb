```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[2,4,3,6,4]

plt.plot(x,y,color="r")
plt.savefig("Plot1",dpi=2000)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[2,4,3,6,4]

plt.plot(x,y,color="r")
plt.savefig("Plot2",dpi=2000,facecolor="y")
plt.show()
```

⬆ It automatically save pictures in png format, if you want to save it in ijpeg,.pdf,etc than you can simply type as "filename_extension"       

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[2,4,3,6,4]

plt.plot(x,y,color="r")
plt.savefig("Plot3.pdf",dpi=2000,facecolor="y")
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[2,4,3,6,4]

plt.plot(x,y,color="r")
plt.savefig("Plot4",dpi=2000,facecolor="y",transparent=True)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[2,4,3,6,4]

plt.plot(x,y,color="r")
plt.savefig("Plot5",dpi=2000,facecolor="y",transparent=True,bbox_inches="tight")
plt.show()
```
