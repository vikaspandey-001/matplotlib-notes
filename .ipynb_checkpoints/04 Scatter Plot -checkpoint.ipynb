# Scatter Plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]
plt.scatter(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]
plt.xlabel("Day",fontsize=20)
plt.ylabel("Number",fontsize=20)
plt.title("Scatter Plot",fontsize=30)
plt.scatter(x,y,color='r')
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]
c=['r','b','g','y','c','r','g']
plt.xlabel("Day",fontsize=20)
plt.ylabel("Number",fontsize=20)
plt.title("Scatter Plot",fontsize=30)
plt.scatter(x,y,color=c)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]

c=['r','b','g','y','c','r','g']

sizes=[50,80,20,60,100,10,25]

plt.xlabel("Day",fontsize=20)
plt.ylabel("Number",fontsize=20)
plt.title("Scatter Plot",fontsize=30)

plt.scatter(x,y,color=c,s=sizes)
plt.show()
```

<h3><i>Example 01</h3></i>

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]

c=[10,49,30,29,56,20,30]

s=[50,80,20,60,100,10,25]

plt.xlabel("Day",fontsize=20)
plt.ylabel("Number",fontsize=20)
plt.title("Scatter Plot",fontsize=30)

plt.scatter(x,y,color=c,sizes=s,marker="*",cmap="viridis")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]

color=[10,49,30,29,56,20,30]

s=[500,800,200,600,100,150,250]

plt.xlabel("Day",fontsize=20)
plt.ylabel("Number",fontsize=20)
plt.title("Scatter Plot",fontsize=30)

plt.scatter(x,y,c=color,sizes=s,marker="*",cmap="viridis")
plt.show()
```

To get colorbar of color map used type "plt.colorbar()"

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]

color=[10,49,30,29,56,20,30]

s=[500,800,200,600,100,150,250]

plt.xlabel("Day",fontsize=20)
plt.ylabel("Number",fontsize=20)
plt.title("Scatter Plot",fontsize=30)

plt.scatter(x,y,c=color,sizes=s,marker="*",cmap="BrBG")

plt.colorbar()

plt.show()
```

Follow to type "colorbar" in horizontal side of colorbar....

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]

color=[10,49,30,29,56,20,30]

s=[500,800,200,600,100,150,250]

plt.xlabel("Day",fontsize=20)
plt.ylabel("Number",fontsize=20)
plt.title("Scatter Plot",fontsize=30)

plt.scatter(x,y,c=color,sizes=s,marker="*",cmap="Set3")

t=plt.colorbar()
t.set_label("Colorbar",fontsize=20)

plt.show()
```
