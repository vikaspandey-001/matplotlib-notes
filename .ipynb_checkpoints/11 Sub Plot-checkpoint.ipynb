```python
import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[1,2,3,4]
plt.plot(x,y,color="r")

plt.pie([1],colors="b")

plt.pie([10,20,30,40])

plt.bar(["a","s","d","f"],[10,20,30,40])

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[1,2,3,4]
plt.subplot(2,2,1)
plt.plot(x,y,color="r")

plt.subplot(2,2,2)
plt.pie([1],colors="b")

plt.subplot(2,2,3)
plt.pie([10,20,30,40])

plt.subplot(2,2,4)
plt.bar(["a","s","d","f"],[10,20,30,40])

plt.show()
```

In all these subplot syntax "plt.subplot(nrows,ncolumns,loc)"
