We can represent line by various symbol (like- ":","--","-.") by using "linefmt" parameter.

##### Example : 1

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]

plt.xlabel("x-axis",fontsize=15,color='m')
plt.ylabel("y-axis",fontsize=15,color='m')
plt.title("Stem Plot",fontsize=25,color='g')

plt.stem(x,y,linefmt=":")
plt.show()
```

To change your marker shape use "markerfmt" parameter

##### Example : 1

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]

plt.xlabel("x-axis",fontsize=15,color='m')
plt.ylabel("y-axis",fontsize=15,color='m')
plt.title("Stem Plot",fontsize=25,color='g')

plt.stem(x,y,linefmt="-.",markerfmt="r+")
plt.show()
```

TO decide fromo which point you want to start your bottom line use "bottom" parameter.

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]

plt.xlabel("x-axis",fontsize=15,color='m')
plt.ylabel("y-axis",fontsize=15,color='m')
plt.title("Stem Plot",fontsize=25,color='g')

plt.stem(x,y,linefmt="-.",markerfmt="ro",bottom=3.5)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]

plt.xlabel("x-axis",fontsize=15,color='m')
plt.ylabel("y-axis",fontsize=15,color='m')
plt.title("Stem Plot",fontsize=25,color='g')

plt.stem(x,y,linefmt="-.",markerfmt="ro",bottom=3.5,basefmt="y")   # it will change the base line color.  
plt.show()
```

To add label in your graph use "label" parameter along with plt.legend().

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]

plt.xlabel("x-axis",fontsize=15,color='m')
plt.ylabel("y-axis",fontsize=15,color='m')
plt.title("Stem Plot",fontsize=25,color='g')

plt.stem(x,y,linefmt="-.",markerfmt="ro",bottom=3.5,basefmt="y",label="Python")    line color.  
plt.legend()
plt.show()
```

You can change location of legend by passing "loc" parameter in legend,
Legend range lies between 0 to 10.


```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]

plt.xlabel("x-axis",fontsize=15,color='m')
plt.ylabel("y-axis",fontsize=15,color='m')
plt.title("Stem Plot",fontsize=25,color='g')

plt.stem(x,y,linefmt="-.",markerfmt="ro",bottom=3.5,basefmt="y",label="Python")     
plt.legend(loc=2)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]

plt.xlabel("x-axis",fontsize=15,color='m')
plt.ylabel("y-axis",fontsize=15,color='m')
plt.title("Stem Plot",fontsize=25,color='g')

plt.stem(x,y,linefmt="-.",markerfmt="ro",bottom=3.5,basefmt="y",label="Python",use_line_collection=True)     
# no change takes place because it defaulted set as True.
plt.legend(loc=2)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]

plt.xlabel("x-axis",fontsize=15,color='m')
plt.ylabel("y-axis",fontsize=15,color='m')
plt.title("Stem Plot",fontsize=25,color='g')

plt.stem(x,y,linefmt="-.",markerfmt="ro",bottom=3.5,basefmt="y",label="Python",use_line_collection=False)     
# you will observe that color of all lines changes.
plt.legend(loc=2)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]

plt.xlabel("x-axis",fontsize=15,color='m')
plt.ylabel("y-axis",fontsize=15,color='m')
plt.title("Stem Plot",fontsize=25,color='g')

plt.stem(x,y,linefmt="-.",markerfmt="ro",bottom=3.5,basefmt="y",label="Python",use_line_collection=False,
        orientation="horizontal")     
plt.legend(loc=2)
plt.show()
```
