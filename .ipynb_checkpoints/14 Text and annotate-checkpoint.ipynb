# Text

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[3,1,5,2,6]

plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[3,1,5,2,6]

plt.xlabel("x-axis",fontsize=15)
plt.ylabel("y-axis",fontsize=15)
plt.title("Text",fontsize=25)
plt.plot(x,y)
plt.text(3,5,"Python",fontsize=10,color="r",style="italic")
plt.show()
```

# Annotate

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[3,1,5,2,6]

plt.xlabel("x-axis",fontsize=15)
plt.ylabel("y-axis",fontsize=15)
plt.title("Text",fontsize=25)

plt.plot(x,y)
plt.text(3,5,"Python",fontsize=10,color="r",style="italic",bbox={"facecolor":"y"})
plt.annotate("Java",xy=(2,1),xytext=(3,2),arrowprops={"facecolor":"pink"})

leg=plt.legend(["Python"],loc=9,facecolor="black",edgecolor="yellow",framealpha=0.5,shadow=True)
for text in leg.get_texts():
    text.set_color("green")

plt.show()
```
