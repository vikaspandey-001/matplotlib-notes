```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
area1=[2,3,2,5,4]
area2=[2,3,4,5,6]
area3=[1,3,2,4,2]

l=["area1","area2","area3"]

plt.xlabel("x-axis",fontsize=20)
plt.ylabel("y-axis",fontsize=20)
plt.title("Stack Plot",fontsize=30)

plt.stackplot(x,area1,area2,area3)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
area1=[2,3,2,5,4]
area2=[2,3,4,5,6]
area3=[1,3,2,4,2]

l=["area1","area2","area3"]

plt.xlabel("x-axis",fontsize=20)
plt.ylabel("y-axis",fontsize=20)
plt.title("Stack Plot",fontsize=30)

plt.stackplot(x,area1,area2,area3,labels=l)
plt.legend(loc=2)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
area1=[2,3,2,5,4]
area2=[2,3,4,5,6]
area3=[1,3,2,4,2]

l=["area1","area2","area3"]

plt.xlabel("x-axis",fontsize=20)
plt.ylabel("y-axis",fontsize=20)
plt.title("Stack Plot",fontsize=30)

plt.stackplot(x,area1,area2,area3,labels=l,colors=["r","g","b"])
plt.legend(loc=2)
plt.show()
```
