# Pie Chart

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
plt.pie(x)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["C","C++","Java","Python"]
plt.pie(x,labels=y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["C","C++","Java","Python"]
ex=[0.2,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["C","C++","Java","Python"]
ex=[0.2,0.4,0.3,0.1]
plt.pie(x,labels=y,explode=ex)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["C","C++","Java","Python"]
ex=[0.2,0.4,0.3,0.1]
c=["r","b","g","c"]
plt.pie(x,labels=y,explode=ex,colors=c)
plt.show()
```

To get percentage of area in Pie Chart use autopct="%1.1%%".
Here 1.1 represents how many numbers you want after the decimal.

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["C","C++","Java","Python"]
ex=[0.2,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex,autopct="%1.1f%%")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["C","C++","Java","Python"]
ex=[0.2,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex,autopct="%1.2f%%")
plt.title("Pie Chart")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["C","C++","Java","Python"]
ex=[0.2,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex,autopct="%1.2f%%",shadow=True)
plt.title("Pie Chart")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["C","C++","Java","Python"]
ex=[0.2,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex,autopct="%1.2f%%",shadow=True,radius=1.5)      # radius  is used to define radius of pie chart.        
plt.title("Pie Chart")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["C","C++","Java","Python"]
ex=[0.2,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex,autopct="%1.2f%%",shadow=True,radius=1.2,labeldistance=1.075)      # labedistance  is used to determine distance of labels from pie chart.            
plt.title("Pie Chart")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["C","C++","Java","Python"]
ex=[0.2,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex,autopct="%1.2f%%",shadow=True,radius=1.2,labeldistance=1.075,startangle=90)      # startangle is used to determine from which angle you want to start like 0,90,180,270..                   
plt.title("Pie Chart")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["C","C++","Java","Python"]
ex=[0.2,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex,autopct="%1.2f%%",shadow=True,radius=1.2,labeldistance=1.075,startangle=270)      # startangle is used to determine from which angle you want to start like 0,90,180,270..                   
plt.title("Pie Chart")
plt.show() 
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40]
y=["C","C++","Java","Python"]

ex=[0.2,0.0,0.0,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%1.2f%%",shadow=True,radius=1.2,labeldistance=1.075,startangle=270,
       textprops={"fontsize":15})

plt.title("Pie Chart")
plt.show() 
```

Here we can't apply directly changes on font like we did in histogram and bar graph.
We have to pass a dictionary textprops for all changes that we want to do in text.

```python
import matplotlib.pyplot as plt

x=[10,20,30,40]
y=["C","C++","Java","Python"]

ex=[0.2,0.0,0.0,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%1.2f%%",shadow=True,radius=1.2,labeldistance=1.075,startangle=270,
       textprops={"fontsize":15}, counterclock=True)

plt.title("Pie Chart")
plt.show() 
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40]
y=["C","C++","Java","Python"]

ex=[0.2,0.0,0.0,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%1.2f%%",shadow=True,radius=1.2,labeldistance=1.075,startangle=270,
       textprops={"fontsize":15},counterclock=False)

plt.title("Pie Chart")
plt.show() 
```

It will rotate the notation of labels in clockwise or counterxlock wise according to the input.

```python
import matplotlib.pyplot as plt

x=[10,20,30,40]
y=["C","C++","Java","Python"]

ex=[0.2,0.0,0.0,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%1.2f%%",shadow=True,radius=1.2,labeldistance=1.075,startangle=270,
       textprops={"fontsize":15},wedgeprops={"linewidth":16})

plt.title("Pie Chart")
plt.show() 

# You can see the changes in piechart.
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40]
y=["C","C++","Java","Python"]

ex=[0.2,0.0,0.0,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%1.2f%%",shadow=True,radius=1.2,labeldistance=1.075,startangle=270,
       textprops={"fontsize":15},wedgeprops={"linewidth":16,"width":1})

plt.title("Pie Chart")
plt.show() 

# It will increase the width of each section
# You can see the changes in piechart.
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40]
y=["C","C++","Java","Python"]

ex=[0.2,0.0,0.0,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%1.2f%%",shadow=True,radius=1.2,labeldistance=1.075,startangle=270,
       textprops={"fontsize":15},wedgeprops={"linewidth":3,"width":1,"edgecolor":"w"})

plt.title("Pie Chart")
plt.show() 

# You can see the changes in piechart.
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40]
y=["C","C++","Java","Python"]

ex=[0.2,0.0,0.0,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%1.2f%%",shadow=True,radius=1.2,labeldistance=1.075,startangle=270,
       textprops={"fontsize":15},wedgeprops={"linewidth":3,"edgecolor":"w"},
       center=(3,6))   # center will d=shift from (0,0) to (3,6) 

plt.title("Pie Chart")
plt.show() 

# You can see the changes in piechart.
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40]
y=["C","C++","Java","Python"]

ex=[0.2,0.0,0.0,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%1.2f%%",shadow=True,radius=1.2,labeldistance=1.075,startangle=270,
       textprops={"fontsize":15},wedgeprops={"linewidth":16},rotatelabels=True)

plt.title("Pie Chart")
plt.show() 

# You can see the changes in piechart.
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40]
y=["C","C++","Java","Python"]

ex=[0.2,0.0,0.0,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%1.2f%%",shadow=True,radius=1.2,labeldistance=1.075,startangle=270,
       textprops={"fontsize":15},wedgeprops={"linewidth":16},rotatelabels=False)

plt.title("Pie Chart")
plt.legend()   # used to add legend in pie chart as we have already discussed in bar graph and histogram.
plt.show() 

# You can see the changes in piechart.
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40]
y=["C","C++","Java","Python"]

ex=[0.2,0.0,0.0,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%1.2f%%",shadow=True,radius=1.2,labeldistance=1.075,startangle=270,
       textprops={"fontsize":15},wedgeprops={"linewidth":16},rotatelabels=False)

plt.title("Pie Chart")
plt.legend(loc=3)      # "loc" parameter is used to determine that where you want to attach your legend.   
plt.show() 

# You can see the changes in piechart.
```

### Now we are gping to understand that how we can make piechart in a piechart or concentric pie chart.

Before starting lets look into a concept

```python
import matplotlib.pyplot as plt
plt.pie([1])
plt.show()
```

What is does ??
It creates a simple pie chart with radius 1 unit

Lets first create a simple pie chart

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["C","C++","Java","Python"]
plt.pie(x,labels=y)
plt.show()
```

Now, we have a parameter called radius used to define the radius of pie chart. Than what we are going to do is to create two piechart one having large radius and another having small radius.

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["C","C++","Java","Python"]

x1=[40,30,20,10]

plt.pie(x,labels=y,radius=1.5)
plt.pie(x1,labels=y,radius=1)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["C","C++","Java","Python"]

plt.pie(x,labels=y,radius=1.5)
plt.pie([1],colors="w")     # It will create a white colored pie chart with 1 unit radius. Which than after looks like ring.       
plt.show()
```
