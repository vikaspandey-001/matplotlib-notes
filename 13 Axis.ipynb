##### We use xticks an yticks to custom decide the ticks on axes.

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[3,1,5,2,4]

plt.plot(x,y)
plt.xticks(x)
plt.yticks(y)
plt.show()
```

To get custom ticks pass a list within xticks and ytick.

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[3,1,5,2,4]

plt.plot(x,y)
plt.xticks(x,labels=["C","R","Java","C++","Python"])
plt.yticks(y)
plt.show()
```

##### xlim and ylim is used to decide the limit of x-axis and y-axis

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[3,1,5,2,4]

plt.plot(x,y)
plt.xlim(0,10)
plt.ylim(0,10)
plt.show()
```
