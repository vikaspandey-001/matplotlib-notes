```python
import matplotlib.pyplot as plt
import numpy as np

x=['Python','C','C++','Java']
y=[85,70,60,82]
a=[20,30,40,50]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel('language',fontsize=30)
plt.ylabel('No.',fontsize=20)
plt.title('Data',fontsize=20)

plt.bar(p,y,width,color='c',align='center',edgecolor='black',label="My Preference")
plt.bar(p1,a,width,color='black',align='center',edgecolor='black',label="Others Preference")

plt.legend()
plt.show()
```

but here the problem is: it shows index number of x on x-axis. To solve this problem look at next step.

```python
import matplotlib.pyplot as plt
import numpy as np

x=['Python','C','C++','Java']
y=[85,70,60,82]
a=[20,30,40,50]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel('language',fontsize=30)
plt.ylabel('No.',fontsize=20)
plt.title('Data',fontsize=20)

plt.bar(p,y,width,color='c',align='center',edgecolor='black',label="My Preference")
plt.bar(p1,a,width,color='black',align='center',edgecolor='black',label="Others Preference")

plt.xticks(p+width,x)
plt.legend()
plt.show()
```

to show x label in centre we divide p+width by 2

```python
import matplotlib.pyplot as plt
import numpy as np

x=['Python','C','C++','Java']
y=[85,70,60,82]
a=[20,30,40,50]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel('language',fontsize=30)
plt.ylabel('No.',fontsize=20)
plt.title('Data',fontsize=20)

plt.bar(p,y,width,color='c',align='center',edgecolor='black',label="My Preference")
plt.bar(p1,a,width,color='black',align='center',edgecolor='black',label="Others Preference")

plt.xticks(p+width/2,x)
plt.legend()
plt.show()
```

to rotate x axis level we use rotation argument. 

```python
import matplotlib.pyplot as plt
import numpy as np

x=['Python','C','C++','Java']
y=[85,70,60,82]
a=[20,30,40,50]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel('language',fontsize=30)
plt.ylabel('No.',fontsize=20)
plt.title('Data',fontsize=20)

plt.bar(p,y,width,color='c',align='center',edgecolor='black',label="My Preference")
plt.bar(p1,a,width,color='black',align='center',edgecolor='black',label="Others Preference")

plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```
