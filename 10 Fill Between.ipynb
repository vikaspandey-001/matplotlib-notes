```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.xlabel("x-axis",fontsize=20)
plt.ylabel("y-axis",fontsize=20)
plt.title("Python",fontsize=30)

plt.plot(x,y,color="r")
plt.fill_between(x,y,color="c",where=(x>=2) & (x<=4))
plt.show()
```

 It shows an error because in matplotlib '>=' not supported between instances of 'list' and 'int'.
It supports number in form of array.
###### So we have to provide a numpy as array to fulfill its condition.

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])

plt.xlabel("x-axis",fontsize=20)
plt.ylabel("y-axis",fontsize=20)
plt.title("Python",fontsize=30)

plt.plot(x,y,color="r")
plt.fill_between(x,y,color="c",where=(x>=2) & (x<=4))
plt.show()
```
