```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70,120]
x1=[10,20,40,50,30,60,90]

plt.boxplot(x,labels=["Python"],showmeans=True,sym="b+",
           boxprops={"color":"r"},capprops={"color":"c"},whiskerprops={"color":"g"})
plt.boxplot(x1,labels=["Anaconda"],showmeans=True,sym="b+",
           boxprops={"color":"r"},capprops={"color":"c"},whiskerprops={"color":"g"})

plt.show()
```

⬆ By doing this it will overlaped on each other.

⬇ To separate this we have to first pass it in a list

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70]
x1=[10,20,40,50,30,60,90]

y=[x,x1]

plt.boxplot(y,labels=["Python","Anaconda"],showmeans=True,sym="b+",
           boxprops={"color":"r"},capprops={"color":"c"},whiskerprops={"color":"g"})

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70]
x1=[10,20,40,50,30,60,90]
x2=[20,40,60,80,100,150,200]

y=[x,x1,x2]

plt.boxplot(y,labels=["Python","Anaconda","King Cobra"],showmeans=True,sym="b+",
           boxprops={"color":"r"},capprops={"color":"c"},whiskerprops={"color":"g"})

plt.show()
```
